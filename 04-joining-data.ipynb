### joining-data

This is just a quick exercise to try and join the collection data with the pen data using pandas.

```python
## read in the data as a pandas object
import pandas as pd
pen_df = pd.read_csv('../the-pen-data/data/pen-collected-items.csv')

## show the first 10 rows
pen_df[:10]
```

```python
## read in the collection data as another panda object
collection_df = pd.read_csv('../collection/meta/objects.csv', low_memory=False)

collection_df[:5]
```

```python
## join the two data sets using pandas merge function 
result = pd.merge(pen_df, collection_df, left_on='refers_to_object_id', right_on='id')
```
